Delete Unamed column and Rank

Some of the players names in players.csv have asterisk, have to get rid of that

Greg Anderson has multiple rows associated to his name, we want to have his name only once
He has multiple rows because he played for multiple teams in 1991

```python
players.groupby(['Player','Year']).get_group(("Greg Anderson",1991))
```

Create function to iterate through each group so that there is only one.
If there is more than one name then we take the row with the team TOT as that is the Total of all the team he played for.
However, TOT is not a team name so replace TOT with last team he played with (last index)

```python
def one_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm']== "TOT"]
        #replace TOT with last team name
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row
players = players.groupby(['Player','Year']).apply(one_row)
```

```python
players.head(20)
```

Two levels of index were create (Player and Year) we can get rid of these indexes to make it easier to work with database

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

We can now merge the two data frames

```python
combined = players.merge(mvps, how='outer', on = ['Player','Year'])
#do outer merge because not all rows in players are in mvp dataframe
```

Replace the people with Na with zeros in their data

```python
combined[['Pts Won', 'Pts Max', 'Share']]= combined[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

Read the Teams Data

```python
teams.head(30)
```

Some rows have divison instead of numerical value
Have to get rid of this by using pandas contain

```python
teams = teams[~teams['W'].str.contains('Division')]
#if the column in W has the word Division in it, get rid of that row
```

```python
teams.head(30)
```

```python
teams['Team'] = teams['Team'].str.replace('*',"",regex=False)
```

```python
teams.head()
```

The team names in Teams Dataframe are the complete team name, whereas team in Combined dataframe are abreviations

Create a nicknames list

```python
nicknames = {}
with open('nicknames.csv') as f:
    lines = f.readlines()
    for line in lines [1:]: #skip first row because its a header
        abbrev,name = line.replace('\n',"").split(',') #replace the '\n' with empty column and split at the ','
        nicknames[abbrev] = name
```

```python
nicknames #test nicknames
```

```python
combined["Team"] = combined['Tm'].map(nicknames)
```

```python
stats = combined.merge(teams, how='outer', on=["Team","Year"])
```
